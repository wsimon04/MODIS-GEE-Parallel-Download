```python
#!/usr/bin/env python3
import subprocess

START_DATE = "2000-01-01"  # specify desired start date
END_DATE = "2025-12-31"    # specify desired end date
NUM_GROUPS = 5             #specify number of groups to match number of parquet files

JOB_SCRIPT = "/insert/path/here"  # path to run job .sh script

print(f"Number of groups: {NUM_GROUPS} for {START_DATE} -> {END_DATE}")

for group in range(1, NUM_GROUPS + 1):
    print(f"Submitting group {group}")
    subprocess.run([
        "sbatch",
        f"--export=ALL,GROUP_ID={group},START_DATE={START_DATE},END_DATE={END_DATE}",
        JOB_SCRIPT
    ])
```
